```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
pic=cv2.imread("husky1.jfif")
pic=cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
plt.axis("off")
plt.imshow(pic)
plt.show()
r,c,d=pic.shape
M = np.float32([[1,0,50],
               [0,1,75],
               [0,0,1]])
translated_img = cv2.warpPerspective(pic,M,(c,r))
plt.axis("off")
plt.imshow(translated_img)
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
pic=cv2.imread("husky1.jfif")
pic=cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
plt.axis("off")
plt.imshow(pic)
plt.show()
r,c,d=pic.shape
M_x = np.float32([[1,0,0],
                 [0,-1,r],
                 [0,0,1]])
M_y = np.float32([[-1,0,c],
                 [0,1,0],
                 [0,0,1]])
reflected_img_x = cv2.warpPerspective(pic,M_x,(int(c),int(r)))
reflected_img_y = cv2.warpPerspective(pic,M_y,(int(c),int(r)))
plt.axis("off")
plt.imshow(reflected_img_x)
plt.show()
plt.axis("off")
plt.imshow(reflected_img_y)
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
pic=cv2.imread("husky1.jfif")
pic=cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
plt.axis("off")
plt.imshow(pic)
plt.show()
r,c,d=pic.shape
angle = np.radians(25)
M = np.float32([[np.cos(angle),-(np.sin(angle)),0],
                [np.sin(angle),np.cos(angle),0],
                [0,0,1]])
rotated_img = cv2.warpPerspective(pic,M,(int(c),int(r)))
plt.axis("off")
plt.imshow(rotated_img)
plt.show()
```
